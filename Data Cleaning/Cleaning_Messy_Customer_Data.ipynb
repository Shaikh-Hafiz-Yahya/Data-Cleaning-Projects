# Technique Data Cleaning

```python
import pandas as pd
```

# load csv = messy_customer_data

```python
df = pd.read_csv(r'C:\Users\Muhammad Yahya\Downloads\python\Pandas\Data Cleaning\messy_customer_data.csv')
print('messy_customer_data')
df
```

# Check Missing Values:

```python
df.isnull().sum()
```

# Spent_Amount: Missing values ko mean() sy fill

```python
df['Spent_Amount'].isnull()
```

```python
df.loc[2 , 'Spent_Amount'] = df['Spent_Amount'].mean()
df
```

# City: karachi (small) ko Karachi (Capital) banana hai.

```python
df['City'] = df['City'].str.capitalize()
df
```

# Name:  Usman  ke agay peeche ki faltu spaces strip() karni hain.

```python
df.loc[5 , 'Name']
```

```python
df.loc[5 , 'Name'] = df.loc[5 , 'Name'].strip()
df
```

```python
df.loc[5,'Name']
```

# Last Name = NAN ---> fill 0

```python

df['Name'] = df['Name'].fillna(0)
df
```

# Drop Duplicated

```python
df.drop_duplicates(inplace=True)
df
```

# Outlier Handle

```python
df.loc[df['Age'] > 100, 'Age'] = 28
```

```python
print('After_Cleaning_Messy_Customer_Data')
df
```
